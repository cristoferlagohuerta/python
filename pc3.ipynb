```python
import operator


OPERATORS = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
}

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def build_expression_tree():
    """1. Construir el árbol con nodos"""
    node_7 = Node(7)
    node_3 = Node(3)
    node_5 = Node(5)
    node_2 = Node(2)

    node_plus = Node('+')
    node_minus = Node('-')
    node_mul = Node('*')

    node_plus.left = node_7
    node_plus.right = node_3

    node_minus.left = node_5
    node_minus.right = node_2

    node_mul.left = node_plus
    node_mul.right = node_minus

    return node_mul

def evaluate_expression_tree(root):
    """2. Evaluar la expresión y mostrar el resultado"""
    if str(root.value) not in OPERATORS:
        return root.value

    op_func = OPERATORS[root.value]
    left_result = evaluate_expression_tree(root.left)
    right_result = evaluate_expression_tree(root.right)

    return op_func(left_result, right_result)

def preorder_traversal(root, result_list):
    """3. Mostrar los tres recorridos: PreOrden"""
    if root:
        result_list.append(str(root.value))
        preorder_traversal(root.left, result_list)
        preorder_traversal(root.right, result_list)

def inorder_traversal(root, result_list):
    """3. Mostrar los tres recorridos: InOrden"""
    if root:
        inorder_traversal(root.left, result_list)
        result_list.append(str(root.value))
        inorder_traversal(root.right, result_list)

def postorder_traversal(root, result_list):
    """3. Mostrar los tres recorridos: PostOrden"""
    if root:
        postorder_traversal(root.left, result_list)
        postorder_traversal(root.right, result_list)
        result_list.append(str(root.value))


expression_tree_root = build_expression_tree()


result = evaluate_expression_tree(expression_tree_root)


preorder_list = []
preorder_traversal(expression_tree_root, preorder_list)

inorder_list = []
inorder_traversal(expression_tree_root, inorder_list)

postorder_list = []
postorder_traversal(expression_tree_root, postorder_list)



print("=== ÁRBOL DE EXPRESIÓN ===")
print("Expresión: (7 + 3) * (5 - 2)")
print(f"Resultado: {result}")
print()
print(f"Recorrido PreOrden: {' '.join(preorder_list)}")
print(f"Recorrido InOrden: {' '.join(inorder_list)}")
print(f"Recorrido PostOrden: {' '.join(postorder_list)}")
```
