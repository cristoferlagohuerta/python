```python
#un ABB es como un arbol genealogico 
# pero con numeros 
#1. cada nodo tiene como maximo  2 hijos 
#2. izquierda: numeros menores 
#3. derecha : numeros mayores 

"""
                10
               /  \
              5   15 
             / \  / \
            3  7  12 20
"""
```

```python
# leccion 1: creando un nodo 
#un nodo es como una caja 
# guarda: el valor(numero)
# guarda: dos fechas apuntando a sus hijos

class Nodo:
    def __init__(self,valor):
        self.valor = valor
        self.izq = None # Hijo izquierdo (numero menores)
        self.der = None # Hijo derecho (numeros mayores)
# creamo un nodo simple 
raiz = Nodo (10)
print(f" Nodo creado con valor : {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izq}")
print(f"hijo derecho: {raiz.der}")




```
